```python
url = 'https://maps.roadtrippers.com/?lng=-124.54651168458969&lat=48.94490114013851&z=8.061000386283393&a2=p!vp*5*9*14*16,s!108*109*115*106*105*113*114*104*112*110'
path_chrome = r"D:\chromedriver"

driver = webdriver.Chrome(path_chrome)
driver.get(url)
time.sleep(10)
print("{} got loaded".format(url))
```

```python
data = {'Attractions' : []}
data
```

```python
soup = bs(driver.page_source, 'html.parser')
soup
```

```python
res = soup.find_all('li', class_='place-list-item')
len(res)
```

```python
for i in range(len(res)):
    try:
        s = res[i].text.split('.')[1]
    except:
        s = res[i].text
    s = "".join(filter(lambda x: not x.isdigit(), s))
    if data['Attractions'].count(s) == 0:
        data['Attractions'].append(s)
```
